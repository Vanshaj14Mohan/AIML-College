```python
from datetime import date
```

```python
df["Cars_Age"] = date.today().year-df["Year"]
df.head()
```

```python
labels = ["Recent", "Old", "Very Old"]
bins = [0,5,10,20]
df["Car_Bin"] = pd.cut(df["Cars_Age"], bins, labels=labels,)
df.head()
```

```python
recent_cars = df[df["Car_Bin"] == "Recent"]
print(recent_cars)
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(8, 5))
sns.countplot(x="Car_Bin", data=df, order=["Recent", "Old", "Very Old"],
              palette={"Recent": "green", "Old": "orange", "Very Old": "red"})

plt.title("Number of Cars by Age Category", fontsize=14)
plt.xlabel("Car Age Category", fontsize=12)
plt.ylabel("Number of Cars", fontsize=12)
plt.figure(figsize=(10,6))

plt.tight_layout()
plt.show()
```
